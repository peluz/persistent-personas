```python
import json
from transformers import AutoTokenizer
import pandas as pd
```

```python
roles = json.load(open("./data/roles.json", "r"))
```

```python
roles.keys()
```

```python
dialogues = {}

for role in roles.keys():
    dialogues[role] = [json.load(open(f"./generations/gemma-3-27b-it/interview_{role.replace(" ", "_")}.json", "r"))]
```

```python
model_id = "google/gemma-3-27b-it"
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
dialogues_df = pd.DataFrame.from_dict(dialogues, orient="index").rename(columns=lambda x: "dialogue")
```

```python
dialogues_df["turns"] = dialogues_df["dialogue"].apply(lambda x: len(x)/2)
```

```python
dialogues_df["length"] = dialogues_df["dialogue"].apply(lambda x: len(tokenizer.apply_chat_template(x, tokenize=True)))
```
